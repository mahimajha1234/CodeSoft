# Import Libraries:-

```python
import pandas as pd 
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("creditcard.csv")
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

# Data Visualization:-

```python
plt.figure(figsize = (6,4))
sns.histplot(df['Amount'], bins = 50, log_scale = True , kde = True , color = 'green')
plt.title("Transaction Amount Distribution")
plt.show()
```

```python
plt.figure(figsize = (6,4))
sns.histplot(df['Time'], bins = 50 , kde=True , color='orange')
plt.title("Time Distribution")
plt.show()
```

```python
plt.figure(figsize = (6,4))
sns.boxplot(x ='Class', y ='Amount', data = df)
plt.yscale('log')
plt.title("Amount vs Class")
plt.show()
```

```python
df['Amount_log'] = np.log1p(df['Amount'])
```

```python
df['Hour'] = (df['Time'] // 3600) % 24
```

# Train-Test-Split:-

```python
x = df.drop('Class', axis=1)
y = df['Class']
```

```python
from sklearn.model_selection import train_test_split, StratifiedKFold
```

```python
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size = 0.2, stratify = y, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
```

```python
x_train[['Time','Amount','Amount_log','Hour']] = scale.fit_transform(x_train[['Time','Amount','Amount_log','Hour']])
x_test[['Time','Amount','Amount_log','Hour']] = scale.transform(x_test[['Time','Amount','Amount_log','Hour']])
```

# KNeighborsClassifier:-

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

```python
knn.fit(x_train, y_train)
```

```python
knn_pred = knn.predict(x_test)
knn_pred
```

```python
from sklearn.metrics import classification_report, roc_auc_score, average_precision_score, confusion_matrix
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
knn_confusion = confusion_matrix(y_test, knn_pred)
sns.heatmap(knn_confusion, annot = True, fmt = 'd', cmap = 'Blues')
plt.title("KNN Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
print("KNN Classification Report:")
print(classification_report(y_test, knn_pred))
```

```python
knn_acc = accuracy_score(y_test, knn_pred)
knn_per = precision_score(y_test, knn_pred)
knn_rec = recall_score(y_test, knn_pred)
print(f"KNN Accuracy: {knn_acc:.4f}")
print(f"KNN Precision: {knn_per:.4f}")
print(f"KNN Recall: {knn_rec:.4f}")
```

```python
# Check Overfitting
knn_train_acc = accuracy_score(y_train, knn.predict(x_train))
print(f"KNN Training Accuracy: {knn_train_acc:.4f}")
knn_test_acc = accuracy_score(y_test, knn.predict(x_test))
print(f"KNN Test Accuracy: {knn_test_acc:.4f}")
```
